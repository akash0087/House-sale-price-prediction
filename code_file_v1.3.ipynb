```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv(r'C:\Users\akash\Desktop\HackerEarth\HackerEarth Machine Learning Challenge - Carnival Wars!\dataset\train.csv')
df_test = pd.read_csv(r'C:\Users\akash\Desktop\HackerEarth\HackerEarth Machine Learning Challenge - Carnival Wars!\dataset\test.csv')

df_train.head(3)
```

```python
print("Null value check for training data set:")
print(df_train.isnull().sum())

print("\nNull value check for testing data set:")
print(df_test.isnull().sum())
```

```python
df_train_copy = df_train.copy()
df_test_copy = df_test.copy()

#df_copy.head(10)

#df_train_copy.drop(['Product_id', 'Stall_no', 'instock_date', 'Customer_name'], axis = 1, inplace = True)
df_train_copy.drop(['Customer_name'], axis = 1, inplace = True)
#print(df_train_copy.head(10))

#df_test_copy.drop(['Product_id', 'Stall_no', 'instock_date', 'Customer_name'], axis = 1, inplace = True)
df_test_copy.drop(['Customer_name'], axis = 1, inplace = True)
#print(df_test_copy.head(10))
```

```python
#### Missing value treatment using Median & mode ####

print("Null value check for training data set before pre processing:")
print(df_train_copy.isnull().sum())

print("\nNull value check for testing data set before pre processing:")
print(df_test_copy.isnull().sum())

if (df_train_copy.isnull().sum().sum() > 0):
    #print("Inside df_train_copy if stmt...")
    df_train_copy['Stall_no'].fillna(df_train_copy['Stall_no'].mode()[0], inplace=True)
    df_train_copy['Discount_avail'].fillna(df_train_copy['Discount_avail'].mode()[0], inplace=True)
    df_train_copy['charges_1'].fillna(df_train_copy['charges_1'].mode()[0], inplace=True)
    df_train_copy['charges_2 (%)'].fillna(df_train_copy['charges_2 (%)'].mode()[0], inplace=True)
    df_train_copy['Minimum_price'].fillna(df_train_copy['Minimum_price'].median(), inplace=True)
    df_train_copy['Maximum_price'].fillna(df_train_copy['Maximum_price'].median(), inplace=True)
    df_train_copy['Selling_Price'].fillna(df_train_copy['Selling_Price'].median(), inplace=True)

if (df_test_copy.isnull().sum().sum() > 0):
    #print("Inside df_test_copy if stmt...")
    df_test_copy['Stall_no'].fillna(df_test_copy['Stall_no'].mode()[0], inplace=True)
    df_test_copy['Discount_avail'].fillna(df_test_copy['Discount_avail'].mode()[0], inplace=True)
    df_test_copy['charges_1'].fillna(df_test_copy['charges_1'].mode()[0], inplace=True)
    df_test_copy['charges_2 (%)'].fillna(df_test_copy['charges_2 (%)'].mode()[0], inplace=True)
    df_test_copy['Minimum_price'].fillna(df_test_copy['Minimum_price'].median(), inplace=True)
    df_test_copy['Maximum_price'].fillna(df_test_copy['Maximum_price'].median(), inplace=True)


print("\nNull value check for training data set after pre processing:")
print(df_train_copy.isnull().sum())

print("\nNull value check for testing data set after pre processing:")
print(df_test_copy.isnull().sum())
```

```python
# To check if the selling price value in training data set is > 0

print("Is there any -ve values in col: Selling_Price: ", any(df_train_copy['Selling_Price']<0))

df_train_copy['Selling_Price'] = abs(df_train_copy['Selling_Price'])

print("Is there any -ve values in col: Selling_Price after treatment: ", any(df_train_copy['Selling_Price']<0))
```

```python
##### Converting the column: Loyalty_customer into numeric categorical variable #####

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_train_copy['Loyalty_customer_conv'] = le.fit_transform(df_train_copy['Loyalty_customer'])
df_train_copy['Product_Category_conv'] = le.fit_transform(df_train_copy['Product_Category'])
df_train_copy['Product_id_conv'] = le.fit_transform(df_train_copy['Product_id'])
df_train_copy['instock_date_conv'] = le.fit_transform(df_train_copy['instock_date'])

df_test_copy['Loyalty_customer_conv'] = le.fit_transform(df_test_copy['Loyalty_customer'])
df_test_copy['Product_Category_conv'] = le.fit_transform(df_test_copy['Product_Category'])
df_test_copy['Product_id_conv'] = le.fit_transform(df_test_copy['Product_id'])
df_test_copy['instock_date_conv'] = le.fit_transform(df_test_copy['instock_date'])

#print(df_train_copy.head(10))
#print("\n", df_test_copy.head(10))
```

```python
df_train_copy.drop(['Loyalty_customer'], axis = 1, inplace = True)  #Dropping the old column: Loyalty_customer from training dataset.
df_train_copy.drop(['Product_Category'], axis = 1, inplace = True)  #Dropping the old column: Product_Category from training dataset.
df_train_copy.drop(['Product_id'], axis = 1, inplace = True)
df_train_copy.drop(['instock_date'], axis = 1, inplace = True)


df_test_copy.drop(['Loyalty_customer'], axis = 1, inplace = True)  #Dropping the old column: Loyalty_customer from test dataset.
df_test_copy.drop(['Product_Category'], axis = 1, inplace = True)  #Dropping the old column: Product_Category from test dataset.
df_test_copy.drop(['Product_id'], axis = 1, inplace = True)
df_test_copy.drop(['instock_date'], axis = 1, inplace = True)

#print(df_train_copy.head(10))
#print("\n", df_test_copy.head(10))
```

```python
df_train_copy['charges_2'] = df_train_copy['charges_2 (%)']/100
df_test_copy['charges_2'] = df_test_copy['charges_2 (%)']/100


df_train_copy.drop(['charges_2 (%)'], axis = 1, inplace = True)
df_test_copy.drop(['charges_2 (%)'], axis = 1, inplace = True)
```

```python
df_train_copy = df_train_copy[[col for col in df_train_copy.columns if (col != 'Selling_Price')] + ['Selling_Price']]
df_train_copy.head(3)
```

```python
# Dividing the columns that are required for predictions

x1 = pd.DataFrame(df_train_copy.iloc[:,0:13])
y1 = pd.DataFrame(df_train_copy.iloc[:,-1])

y1 = y1.astype('int')

x2 = df_test_copy.copy()

#print(x1)
#print("\n",y1)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(x1,y1)

dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x1.columns)

#concat two dataframes for better visualization.

featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```

```python
x1.drop(['Product_Category_conv'], axis = 1, inplace = True)
x1.drop(['Stall_no'], axis = 1, inplace = True)
#x1.drop(['Grade'], axis = 1, inplace = True)
#x1.drop(['Demand'], axis = 1, inplace = True)
#x1.drop(['charges_1'], axis = 1, inplace = True)
#x1.drop(['charges_2'], axis = 1, inplace = True)
#x1.drop(['instock_date_conv'], axis = 1, inplace = True)
x1.drop(['Product_id_conv'], axis = 1, inplace = True)
#x1.drop(['Loyalty_customer_conv'], axis = 1, inplace = True)
#x1.drop(['Discount_avail'], axis = 1, inplace = True)

x2.drop(['Product_Category_conv'], axis = 1, inplace = True)
x2.drop(['Stall_no'], axis = 1, inplace = True)
#x2.drop(['Grade'], axis = 1, inplace = True)
#x2.drop(['Demand'], axis = 1, inplace = True)
#x2.drop(['charges_1'], axis = 1, inplace = True)
#x2.drop(['charges_2'], axis = 1, inplace = True)
#x2.drop(['instock_date_conv'], axis = 1, inplace = True)
x2.drop(['Product_id_conv'], axis = 1, inplace = True)
#x2.drop(['Loyalty_customer_conv'], axis = 1, inplace = True)
#x2.drop(['Discount_avail'], axis = 1, inplace = True)
```

```python
from sklearn.linear_model import LinearRegression

LRegressor = LinearRegression()
LRegressor.fit(x1, y1)

#To retrieve the intercept:
print(LRegressor.intercept_)

#For retrieving the slope:
print(LRegressor.coef_)
```

```python
# Predicting the result

y_pred = LRegressor.predict(x2)

# To compare the actual output values for X_test with the predicted values
#comp_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
#comp_df
```
